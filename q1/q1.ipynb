```python
import matplotlib.pyplot as plt
def main():

    print("Hello, World!")

 

if __name__ == '__main__':

 

    main()
def plotdata(dataset,nameset):
    plt.title('Average Temp in 2022')
    for i,data in enumerate(dataset):
        x=[]
        y=[]
        for row in data:
            if row[2]!='':
                x.append(float(row[0].split('-')[1]))
                y.append(float(row[2]))
        plt.plot(x,y,label=nameset[i])
    plt.ylim([-10,35])
    plt.xlabel('month')
    plt.ylabel('temperatue')
    plt.legend()
    plt.show()
```

```python
import csv
fs=open('seoul1.csv','r',encoding='cp949')
fb=open('busan1.csv','r',encoding='cp949')
fd=open('daejon1.csv','r',encoding='cp949')
fj=open('jeju1.csv','r',encoding='cp949')
fw=open('whole1.csv','r',encoding='cp949')
sdata=csv.reader(fs)
bdata=csv.reader(fb)
ddata=csv.reader(fd)
jdata=csv.reader(fj)
wdata=csv.reader(fw)
dataset=[sdata,bdata,ddata,jdata,wdata]
for i in range(5):
    next(dataset[i])
nameset=['Seoul','Busan','Daejon','Jeju','Nationwide']
plotdata(dataset,nameset)
fs.close()
fb.close()
fd.close()
fj.close()
fw.close()
```

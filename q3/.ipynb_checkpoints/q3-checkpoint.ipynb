```python
import csv

fj=open('jeju3.csv','r',encoding='cp949')
fw=open('nationwide3.csv','r',encoding='utf-8')
dataj=csv.reader(fj)
dataw=csv.reader(fw)
next(dataj)
next(dataw)
```

```python
#fw->6+6x
#13,15,17,19
#6,18,30,42
rowj=next(dataj)
roww=next(dataw)
ratejeju=[rowj[6],rowj[18],rowj[30],rowj[42]]
ratewide=[roww[6],roww[18],roww[30],roww[42]]
print(ratejeju)
print(ratewide)
#13년,15년,17년,19년 순서대로 제주도와 전국의 남녀 성비 비교
if(rowj[6]>roww[6]):
    print('true')
else:
    print('false')
if(rowj[18]>roww[18]):
    print('true')
else:
    print('false')
if(rowj[30]>roww[30]):
    print('true')
else:
    print('false')
if(rowj[42]>roww[42]):
    print('true')
else:
    print('false')
fj.close()
fw.close()
```

```python
import matplotlib.pyplot as plt
plt.title('Jeju and nationwide comparison gender rate at 2013~2019')
plt.plot([2013,2015,2017,2019],ratewide,label='wordwide')
plt.plot([2013,2015,2017,2019],ratejeju,label='jeju')
plt.legend()
plt.show()
```

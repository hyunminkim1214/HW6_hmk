```python
def get_inarr():

  data = pd.read_csv("subway4.csv")
  sorted = (pd.Series(map(int, data[1:].iloc[:, 2])) + pd.Series(map(int, data[1:].iloc[:, 4]))).sort_values(ascending=False)
  indices = sorted.index[:30]

  t_30 = []
  for idx in indices:
    t_30.append(tuple([data.iloc[idx, 1], sorted[idx]]))

  return t_30
```

```python
def get_outarr():

  data = pd.read_csv("subway4.csv")
  sorted = (pd.Series(map(int, data[1:].iloc[:, 3])) + pd.Series(map(int, data[1:].iloc[:, 5]))).sort_values(ascending=False)
  indices = sorted.index[:30]

  t_30 = []
  for idx in indices:
    t_30.append(tuple([data.iloc[idx, 1], sorted[idx]]))

  return t_30
```

```python
def get_allarr():

  data = pd.read_csv("subway4.csv")
  sorted = (pd.Series(map(int, data[1:].iloc[:, 3])) + pd.Series(map(int, data[1:].iloc[:, 5]))+pd.Series(map(int, data[1:].iloc[:, 2])) + pd.Series(map(int, data[1:].iloc[:, 4]))).sort_values(ascending=False)
  indices = sorted.index[:30]

  t_30 = []
  for idx in indices:
    t_30.append(tuple([data.iloc[idx, 1], sorted[idx]]))

  return t_30
```

```python
x1=[]
y1=[]
for a,b in get_inarr():
    x1.append(a)
    y1.append(b)
x2=[]
y2=[]
for a,b in get_outarr():
    x2.append(a)
    y2.append(b)
x3=[]
y3=[]
for a,b in get_allarr():
    x3.append(a)
    y3.append(b)
```

```python
import matplotlib.pyplot as plt
from matplotlib import rc
import seaborn as sns

rc('font', family='AppleGothic')

plt.rcParams['axes.unicode_minus'] = False
plt.subplot(1,3,1)
plt.xlabel('승차인원')
plt.barh(x1,y1)
plt.subplot(1,3,2)
plt.xlabel('하차인원')
plt.barh(x2,y2)
plt.subplot(1,3,3)
plt.xlabel('승하차인원')
plt.barh(x3,y3)
plt.subplots_adjust(
    wspace=3,  
    hspace=1)
```
